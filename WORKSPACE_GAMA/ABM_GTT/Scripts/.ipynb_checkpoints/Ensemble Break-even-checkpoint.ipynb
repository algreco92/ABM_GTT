```python
%pylab inline
import pandas as pd
```

```python
experiments = [[] for i in range(6)]
percorso = "../models/exports/"
for j in [8,9,10,11,12]:
    for i in range(10):
        try:
            experiments[j-8].append(pd.read_csv(percorso+"batch_"+str(j)+"_controllori/ticket_trend_"+str(i)+"_"+str(j)+".csv",header = False, names = ['Time','1','2','3','4','Abbonati','Utility','Total_risk']))
            experiments[j-8][i]['incremento']=zeros(len(experiments[j-8][i]['Time']))
            shifted = [list(experiments[j-8][i]['Utility'])[0]] + list(experiments[j-8][i]['Utility'])[:-1]
            experiments[j-8][i]['incremento'] = array(list(experiments[j-8][i]['Utility']))- array(shifted)
        except:
            pass
```

```python
medie = [[[]for j in experiments[1][1]['Time']]for k in range(len(experiments))]
stds = [[[]for j in experiments[1][1]['Time']]for k in range(len(experiments))]
```

```python
time_utility = [[[]for j in experiments[1][1]['Time'][1:]]for k in range(len(experiments))]#contiene tutti e 520 i tempi, e per ognuno, il rischio in ogni esperimento
for esp in range(len(experiments)):
    for j in range(len(time_utility[esp])):
        for expo in experiments[esp]:
            time_utility[esp][j].append(expo['incremento'][j+1])
print shape(time_utility)
```

```python
for expo in range(len(experiments)):
    for i in range(len(time_utility[expo])):
        medie[expo][i].append(mean(time_utility[expo][i]))
        stds[expo][i].append(std(time_utility[expo][i]))
```
